```python
begin = data.query('status=="Июль 2021"').index[0]
end = data.query('status=="Август 2021"').index[0]
```

```python
#вопрос 1
data_july =  data[(data.index>begin) & (data.index<end)].copy() 
data_july.loc[data['status']!='ПРОСРОЧЕНО']['sum'].sum()
```

```python
begin1 = data.query('status=="Май 2021"').index[0]
begin2 = data.query('status=="Июнь 2021"').index[0]
begin3 = data.query('status=="Июль 2021"').index[0]
begin4 = data.query('status=="Август 2021"').index[0]
begin5 = data.query('status=="Сентябрь 2021"').index[0]
begin6 = data.query('status=="Октябрь 2021"').index[0]
end = len(data)
data_2 = data.copy()
data_2["month"]=""
data_2.loc[(data_2.index > begin1) & (data_2.index < begin2), 'month'] = 'Май'
data_2.loc[(data_2.index > begin2) & (data_2.index < begin3), 'month'] = 'Июнь'
data_2.loc[(data_2.index > begin3) & (data_2.index < begin4), 'month'] = 'Июль'
data_2.loc[(data_2.index > begin4) & (data_2.index < begin5), 'month'] = 'Август'
data_2.loc[(data_2.index > begin5) & (data_2.index < begin6), 'month'] = 'Сентябрь'
data_2.loc[(data_2.index > begin6) & (data_2.index < end), 'month'] = 'Октябрь'
data_2
```

```python
data_graph = data_2.groupby('month')['sum'].agg(['sum'])
data_graph
```

```python
#вопрос 2
from matplotlib.ticker import MultipleLocator
ax = sns.barplot(x=list(data_graph.index)[1:], y=list(data_graph['sum'])[1:], data=data_graph)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
ax.yaxis.set_major_locator(MultipleLocator(base=100000))
plt.show()
```

```python
#задание
data_task = data.copy()
begin1 = data.query('status=="Май 2021"').index[0]
end1 = data.query('status=="Июнь 2021"').index[0]
end2 = data.query('status=="Июль 2021"').index[0]
data_task["month"] = ""
data_task.loc[(data_task.index > begin1) & (data_task.index < end1), 'month'] = '05'
data_task.loc[(data_task.index > end1) & (data_task.index < end2),'month'] = '06'
data_task = data_task.head(end2)
data_task = data_task.loc[((data_task['new/current'] == "новая") & (data_task['status'] == "ОПЛАЧЕНО") & (data_task['document'] == "оригинал")) | ((data_task['new/current'] == "текущая") & (data_task['status']!="ПРОСРОЧЕНО") & (data_task['document']=="оригинал"))]
data_task.loc[(data_task['new/current']=="новая") & (data_task['status']=="ОПЛАЧЕНО") & (data_task['document']=="оригинал"),'sum']=data_task.loc[(data_task['new/current']=="новая") & (data_task['status']=="ОПЛАЧЕНО") & (data_task['document']=="оригинал"),'sum']*0.07
data_task.loc[(data_task['new/current']=="текущая") & (data_task['status']!="ПРОСРОЧЕНО") & (data_task['document']=="оригинал") & (data_task['sum']<10000),'sum']=data_task.loc[(data_task['new/current']=="текущая") & (data_task['status']!="ПРОСРОЧЕНО") & (data_task['document']=="оригинал") & (data_task['sum']<10000),'sum']*0.03
data_task.loc[(data_task['new/current']=="текущая") & (data_task['status']!="ПРОСРОЧЕНО") & (data_task['document']=="оригинал") & (data_task['sum']>=10000),'sum']=data_task.loc[(data_task['new/current']=="текущая") & (data_task['status']!="ПРОСРОЧЕНО") & (data_task['document']=="оригинал") & (data_task['sum']>=10000),'sum']*0.05
data_task['receiving_date'] = data_task['receiving_date'].map(lambda x: str(x)[:10])
data_task
```
